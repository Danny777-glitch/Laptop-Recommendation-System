```python
df = pd.read_csv("laptops_Clean.csv")
print(df.shape)
df.head()
```

```python
df = df[[
    "Brand",
    "Processor",
    "GPU",
    "RAM",
    "Storage",
    "Display_Resolution",
    "Refresh_Rate",
    "Price"
]]
df.head(2)
```

```python
def extract_ram(text):
    text = str(text).upper()

    # Case 1: 16GB, 8 GB, 32GB
    match = re.search(r'(\d+)\s*GB', text)
    if match:
        return int(match.group(1))

    # Case 2: 2*8, 2x8
    match = re.search(r'(\d+)\s*[*X]\s*(\d+)', text)
    if match:
        return int(match.group(1)) * int(match.group(2))

    return np.nan
df["RAM_GB"] = df["RAM"].apply(extract_ram)

df[["RAM", "RAM_GB"]].head(10)
```

```python
def extract_storage(text):
    text = str(text).upper()

    total_storage = 0

    tb_matches = re.findall(r'(\d+)\s*TB', text)
    gb_matches = re.findall(r'(\d+)\s*GB', text)

    for tb in tb_matches:
        total_storage += int(tb) * 1024

    for gb in gb_matches:
        total_storage += int(gb)

    return total_storage


df["Storage_GB"] = df["Storage"].apply(extract_storage)

df[["Storage", "Storage_GB"]].head(10)
```

```python
def extract_refresh_rate(text):
    text = str(text).upper()

    match = re.search(r'(\d+)', text)

    if match:
        return int(match.group(1))

    return 60


df["Refresh_Hz"] = df["Refresh_Rate"].apply(extract_refresh_rate)

df[["Refresh_Rate", "Refresh_Hz"]].head(10)
```

```python
cpu_scores = {
    "ULTRA 9": 10,
    "CORE 9": 10,
    "I9": 10,
    "RYZEN 9": 10,

    "ULTRA 7": 8,
    "CORE 7": 8,
    "I7": 8,
    "RYZEN 7": 8,

    "ULTRA 5": 6,
    "CORE 5": 6,
    "I5": 6,
    "RYZEN 5": 6,

    "I3": 4,
    "RYZEN 3": 4,

    "N100": 2,
    "N95": 2,
    "PENTIUM": 2,
    "CELERON": 1
}


def get_cpu_score(cpu):
    cpu = str(cpu).upper()

    for key, score in cpu_scores.items():
        if key in cpu:
            return score

    return 5


df["CPU_Score"] = df["Processor"].apply(get_cpu_score)

df[["Processor", "CPU_Score"]].head(15)
```

```python
def get_gpu_score(gpu):
    gpu = str(gpu).upper()

    match = re.search(r'(RTX|GTX).*?(\d{4})', gpu)

    if match:
        model = int(match.group(2))

        if model >= 5090:
            return 10
        elif model >= 5080:
            return 9.5
        elif model >= 5070:
            return 9
        elif model >= 5060:
            return 8.5
        elif model >= 5050:
            return 8
        elif model >= 4090:
            return 9.5
        elif model >= 4080:
            return 9
        elif model >= 4070:
            return 8.5
        elif model >= 4060:
            return 8
        elif model >= 4050:
            return 7
        elif model >= 3070:
            return 7
        elif model >= 3060:
            return 6.5
        elif model >= 3050:
            return 6
        elif model >= 2050:
            return 5
        elif model >= 1650:
            return 4
        elif model >= 1050:
            return 3

    if "ARC" in gpu:
        return 4

    elif "RADEON" in gpu:
        return 4

    elif "IRIS XE" in gpu:
        return 2

    elif "IRIS" in gpu:
        return 2

    elif "UHD" in gpu:
        return 1.5

    elif "INTEL HD" in gpu:
        return 1

    return 2


df["GPU_Score"] = df["GPU"].apply(get_gpu_score)

df[["GPU", "GPU_Score"]].head(20)
```

```python
def get_resolution_score(resolution):
    resolution = str(resolution).upper()

    match = re.search(r'(\d+)\s*[X*]\s*(\d+)', resolution)

    if match:
        width = int(match.group(1))
        height = int(match.group(2))

        pixels = width * height

        if pixels >= 8000000:      # 4K+
            return 5
        elif pixels >= 4000000:    # QHD+/3K
            return 4
        elif pixels >= 2000000:    # Full HD
            return 3
        elif pixels >= 1000000:    # HD+
            return 2
        else:
            return 1

    return 2


df["Resolution_Score"] = df["Display_Resolution"].apply(get_resolution_score)

df[["Display_Resolution", "Resolution_Score"]].head(20)
```

```python
df["Performance_Score"] = (
    (df["CPU_Score"] / 10 * 35) +
    (df["GPU_Score"] / 10 * 35) +
    (df["RAM_GB"].clip(upper=64) / 64 * 10) +
    (df["Storage_GB"].clip(upper=2048) / 2048 * 10) +
    (df["Resolution_Score"] / 5 * 5) +
    (df["Refresh_Hz"].clip(upper=240) / 240 * 5)
)

df["Performance_Score"] = df["Performance_Score"].round(2)

df[[
    "CPU_Score",
    "GPU_Score",
    "RAM_GB",
    "Storage_GB",
    "Resolution_Score",
    "Refresh_Hz",
    "Performance_Score"
]].head(10)
```

```python
df[["Brand", "Price", "Performance_Score"]].sort_values(
    by="Performance_Score",
    ascending=False
).head(10)
```
